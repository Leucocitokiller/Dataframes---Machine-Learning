```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns



# 2. Seleccionar solo las variables numéricas (presiones y temperaturas)
features = df.select_dtypes(include='number').drop(columns=['Ciclo'])  # dejamos fuera el target si es binario

# 3. Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

# 4. PCA para visualización
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 5. Clustering (k=2 por ahora)
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# 6. Visualización
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='Set2')
plt.title('Clustering KMeans (2 clusters) con PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
import xgboost as xgb

# Paso 1: Convertir timestamp a formato datetime y ordenar datos
df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True)
df = df.sort_values('timestamp')

# Paso 2: Crear la columna que predice si habrá una parada en los próximos X minutos

X_minutos = 10
futuro = df['Ciclo'].shift(-X_minutos)
df['parada_futura'] = (futuro.rolling(window=X_minutos, min_periods=1).min() == 0).astype(int)


# Verificar
print(df[['timestamp', 'Ciclo', 'parada_futura']].tail(20))

# Paso 3: Definir características (X) y target (y)
X = df.drop(columns=['Ciclo', 'parada_futura', 'timestamp'])  # Usar solo variables relevantes
y = df['parada_futura']

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Paso 4: Configurar modelo y pipeline
pipeline_xgb = Pipeline([
    ('clf', xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
])

# Definir hiperparámetros para GridSearch
param_grid_xgb = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [6, 10],
    'clf__learning_rate': [0.1, 0.2],
    'clf__subsample': [0.8, 1.0],
    'clf__colsample_bytree': [0.8, 1.0]
}

# StratifiedKFold para validación cruzada
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Configurar GridSearchCV
grid_xgb = GridSearchCV(pipeline_xgb, param_grid_xgb, cv=skf, scoring='accuracy', n_jobs=-1, verbose=1)

# Paso 5: Entrenar modelo
grid_xgb.fit(X_train, y_train)

# Resultados del GridSearch
print("Mejores parámetros:", grid_xgb.best_params_)
print("Mejor puntuación (validación cruzada):", grid_xgb.best_score_)

# Paso 6: Evaluar modelo en datos de prueba
y_test_pred = grid_xgb.predict(X_test)
y_test_proba = grid_xgb.predict_proba(X_test)[:, 1]  # Probabilidad de clase positiva

# Paso 7: Mostrar métricas de rendimiento
print("📋 Classification Report:")
print(classification_report(y_test, y_test_pred, digits=4))

accuracy = accuracy_score(y_test, y_test_pred)
roc_auc = roc_auc_score(y_test, y_test_proba)
print(f"✅ Accuracy: {accuracy:.4f}")
print(f"✅ ROC AUC: {roc_auc:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
import xgboost as xgb

# Paso 1: Convertir timestamp a formato datetime y ordenar datos
df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True)
df = df.sort_values('timestamp').reset_index(drop=True)

# Paso 2: Detectar inicios de parada (Ciclo pasa de 1 a 0)
df['inicio_parada'] = (df['Ciclo'].shift(1) == 1) & (df['Ciclo'] == 0)

# Paso 3: Crear la columna que indica si en los próximos X minutos habrá una parada
X_minutos = 10  # Cambiar según el horizonte de predicción
df['parada_futura'] = 0

# Identificar los índices de inicio de parada
idx_paradas = df.index[df['inicio_parada'] == True]

# Marcar las X filas anteriores como parada_futura = 1
for idx in idx_paradas:
    start = max(0, idx - X_minutos)
    df.loc[start:idx-1, 'parada_futura'] = 1

# Paso 4: Definir características (X) y variable objetivo (y)
X = df.drop(columns=['Ciclo', 'parada_futura', 'timestamp', 'inicio_parada'])  # Variables de entrada
y = df['parada_futura']  # Target a predecir

# Paso 5: División en train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Paso 6: Definición de pipeline y modelo
pipeline_xgb = Pipeline([
    ('clf', xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
])

# Paso 7: Definición de hiperparámetros
param_grid_xgb = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [6, 10],
    'clf__learning_rate': [0.1, 0.2],
    'clf__subsample': [0.8, 1.0],
    'clf__colsample_bytree': [0.8, 1.0]
}

# Paso 8: Validación cruzada con StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_xgb = GridSearchCV(pipeline_xgb, param_grid_xgb, cv=skf, scoring='accuracy', n_jobs=-1, verbose=1)

# Paso 9: Entrenamiento
grid_xgb.fit(X_train, y_train)

# Paso 10: Evaluación del modelo
y_test_pred = grid_xgb.predict(X_test)
y_test_proba = grid_xgb.predict_proba(X_test)[:, 1]

# Métricas
print("📋 Classification Report:")
print(classification_report(y_test, y_test_pred, digits=4))
accuracy = accuracy_score(y_test, y_test_pred)
roc_auc = roc_auc_score(y_test, y_test_proba)
print(f"✅ Accuracy: {accuracy:.4f}")
print(f"✅ ROC AUC: {roc_auc:.4f}")
```

```python
# Rango de tiempo a mostrar
ventana = 60  # en minutos

# Filtrar datos desde 60 minutos antes hasta el momento de la parada
df_window = df[(df['timestamp'] >= timestamp_parada - pd.Timedelta(minutes=ventana)) &
               (df['timestamp'] <= timestamp_parada)].copy()
```
